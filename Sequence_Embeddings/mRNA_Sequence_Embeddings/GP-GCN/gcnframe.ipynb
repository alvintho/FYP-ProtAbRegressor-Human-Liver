# Use own data to extract embeddings

```python
import torch

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = GCNmodel.model(label_num=2, other_feature_dim=None).to(device)
model_name = "/Users/alvinthosatria/Documents/FYP/datasets/Sequence_Embeddings/mRNA_Sequence_Embeddings/GP-GCN/GCNFrame/GCN_model.pt"
saved_model = torch.load(model_name, map_location=device)
saved_state_dict = saved_model.state_dict()
new_state_dict = model.state_dict()

for name, param in saved_state_dict.items():
    if name in new_state_dict and new_state_dict[name].shape == param.shape:
        new_state_dict[name].copy_(param)
    else:
        print(f"Skipping layer {name} due to shape mismatch or incompatibility.")

model.load_state_dict(new_state_dict, strict=False)

model.eval()
```

```python
from Bio import SeqIO

fasta_file = "/Users/alvinthosatria/Documents/FYP/datasets/Sequence_Embeddings/mRNA_Sequence_Embeddings/GP-GCN/cdna_data/liver_cdna.fasta"
fasta_sequences = list(SeqIO.parse(fasta_file, "fasta"))
total_sequences = len(fasta_sequences)
batch_size = 1551
batches = [fasta_sequences[i:i + batch_size] for i in range(0, total_sequences, batch_size)]
```

```python
import torch
from torch_geometric.data import DataLoader
from GCNFrame import Biodata
import numpy as np
import pandas as pd
from Bio import SeqIO
import os

all_embeddings = []
all_protein_ids = []
all_mrna_ids = []

for batch_idx, batch_sequences in enumerate(batches):
    print(f"\nProcessing batch {batch_idx + 1}/{len(batches)} with {len(batch_sequences)} sequences")
    
    batch_fasta_file = f"batch_{batch_idx}.fasta"
    with open(batch_fasta_file, "w") as f:
        SeqIO.write(batch_sequences, f, "fasta")
    
    data = Biodata(fasta_file=batch_fasta_file)
    batch_dataset = data.encode(thread=20, save_dataset=False)
    print(f"Number of sequences in batch_dataset: {len(batch_dataset)}")
    
    if len(batch_dataset) != len(batch_sequences):
        raise ValueError(f"Batch {batch_idx + 1} has {len(batch_sequences)} sequences, but batch_dataset has {len(batch_dataset)} sequences.")
    
    batch_loader = DataLoader(batch_dataset, batch_size=32, shuffle=False, follow_batch=['x_src', 'x_dst'])
    
    batch_embeddings = []
    with torch.no_grad():
        for batch in batch_loader:
            batch = batch.to(device)
            out, embeddings = model(batch)
            embeddings = embeddings.cpu().numpy()
            batch_embeddings.append(embeddings)
    
    batch_embeddings = np.concatenate(batch_embeddings, axis=0)
    print(f"Shape of batch_embeddings: {batch_embeddings.shape}")
    
    all_embeddings.append(batch_embeddings)
    
    for seq_record in batch_sequences:
        header_parts = seq_record.id.split("|")
        if len(header_parts) >= 2:
            all_protein_ids.append(header_parts[0])
            all_mrna_ids.append(header_parts[1])
        else:
            raise ValueError(f"FASTA header {seq_record.id} does not contain both protein_id and mrna_id")
    
    os.remove(batch_fasta_file)

all_embeddings = np.concatenate(all_embeddings, axis=0)
print(f"Final shape of all_embeddings: {all_embeddings.shape}")  # Should be (7987, 100)

if all_embeddings.shape[0] != total_sequences:
    raise ValueError(f"Expected {total_sequences} embeddings, but got {all_embeddings.shape[0]}.")

embedding_dim = all_embeddings.shape[1]
embedding_columns = [f"emb_{i}" for i in range(embedding_dim)]

df = pd.DataFrame(
    data=np.column_stack([all_protein_ids, all_mrna_ids, all_embeddings]),
    columns=['protein_id', 'mrna_id'] + embedding_columns
)

df[embedding_columns] = df[embedding_columns].apply(pd.to_numeric)

print(df.head())

df.to_csv("liver_cdna_embeddings.csv", index=False)
print("Embeddings DataFrame saved to liver_cdna_embeddings.csv")
```
