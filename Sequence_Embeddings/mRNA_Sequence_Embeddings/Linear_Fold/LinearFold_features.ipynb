# LinearFold Interface

```python
import subprocess
import pandas as pd
from typing import List, Tuple
import concurrent.futures
from tqdm import tqdm


class LinearFold:
    def __init__(
        self, linearfold_path="./LinearFold/linearfold", beam_size=100, model="V"
    ):
        self.linearfold_path = linearfold_path
        self.beam_size = beam_size
        self.model = model  # 'C' for CONTRAfold, 'V' for ViennaRNA

    def predict_structure(self, sequence: str) -> Tuple[str, float]:
        rna_sequence = sequence.replace("T", "U").upper()

        model_flag = f"-{self.model}"

        process = subprocess.Popen(
            [self.linearfold_path, model_flag, "-b", str(self.beam_size)],
            stdin=subprocess.PIPE,
            stdout=subprocess.PIPE,
            stderr=subprocess.PIPE,
            text=True,
        )

        stdout, stderr = process.communicate(input=rna_sequence)
        lines = stdout.strip().split("\n")

        if len(lines) < 2:
            return None, None

        structure = lines[1].split()[0]
        try:
            energy = float(lines[1].split()[-1].strip("()"))  # Extract free energy
        except Exception:
            energy = None

        return structure, energy
```

# Structural Features Functions

```python
def get_structural_features(structure: str) -> dict:
    if structure is None:
        return {
            "paired_ratio": None,
            "unpaired_ratio": None,
        }
    paired = structure.count("(")
    unpaired = structure.count(".")
    total = len(structure)
    return {
        "paired_ratio": paired / total if total > 0 else None,
        "unpaired_ratio": unpaired / total if total > 0 else None,
    }


def process_single_sequence(
    sequence_data: Tuple[str, str], linearfold_path="./LinearFold/linearfold", model="V"
) -> dict:
    seq_id, seq = sequence_data
    lf = LinearFold(linearfold_path, model=model)
    structure, energy = lf.predict_structure(seq)
    features = get_structural_features(structure)
    return {
        "mrna_id": seq_id,
        "energy": energy,  # Free energy in kcal/mol
        **features,
    }


def process_batch(
    sequences: List[str],
    ids: List[str] = None,
    linearfold_path: str = "./LinearFold/linearfold",
    max_workers: int = None,
    model="V",
) -> pd.DataFrame:
    if ids is None:
        ids = [f"seq_{i}" for i in range(len(sequences))]
    sequence_data = list(zip(ids, sequences))
    results = []

    with concurrent.futures.ThreadPoolExecutor(max_workers=max_workers) as executor:
        futures = {
            executor.submit(process_single_sequence, data, linearfold_path, model): data
            for data in sequence_data
        }
        for future in tqdm(
            concurrent.futures.as_completed(futures),
            total=len(futures),
            desc="Processing sequences",
        ):
            results.append(future.result())

    return pd.DataFrame(results)
```

# Region Processor Functions

```python
def process_regions(df, linearfold_path="./LinearFold/linearfold", max_workers=6, model="V"):
    
    regions = ["5utr", "cds", "3utr", "cdna"]
    
    result_df = pd.DataFrame({'mrna_id': df['mrna_id']})
    
    for region in regions:
        print(f"Processing {region} regions...")
        
        valid_indices = df[region].notna()
        region_sequences = df.loc[valid_indices, region].tolist()
        region_mrna_ids = df.loc[valid_indices, 'mrna_id'].tolist()
        
        if not region_sequences:
            print(f"No {region} sequences found.")
            # Add empty columns for this region
            result_df[f"{region}_mfe"] = None
            result_df[f"{region}_paired_ratio"] = None
            result_df[f"{region}_unpaired_ratio"] = None
            continue
        
        region_results = process_batch(
            region_sequences, 
            region_mrna_ids, 
            linearfold_path=linearfold_path, 
            max_workers=max_workers,
            model=model
        )
        
        region_results = region_results.rename(columns={
            'energy': f"{region}_mfe",
            'paired_ratio': f"{region}_paired_ratio",
            'unpaired_ratio': f"{region}_unpaired_ratio"
        })
        
        region_results = region_results[['mrna_id', f"{region}_mfe", f"{region}_paired_ratio", f"{region}_unpaired_ratio"]]
        
        result_df = pd.merge(result_df, region_results, on='mrna_id', how='left')
    
    return result_df

# Usage:
lf_df = process_regions(df, linearfold_path="./LinearFold/linearfold", max_workers=6)
lf_df
```
