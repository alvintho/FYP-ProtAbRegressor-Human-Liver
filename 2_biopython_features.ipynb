```python
import pandas as pd

df = pd.read_csv("/Users/alvinthosatria/Documents/FYP/FYP/cleaned_liver_data_cdna.csv")
df
```

```python
# Convert sequence columns to strings
sequence_columns = ["5utr", "3utr", "cds", "cdna", "protein_sequence"]
for col in sequence_columns:
    df[col] = df[col].astype(str)
```

```python
features_df = pd.DataFrame()
features_df = df[["protein_id", "mrna_id", "protein_length", "cdna_length", "5utr_length", "cds_length", "3utr_length"]]
features_df
```

```python
import sys
import os
sys.path.append(os.path.dirname(os.path.dirname(os.path.abspath(f"{MODULES_PATH}/{FEATURES_CALCULATOR_FILE}"))))
from modules.FeaturesCalculator import MRNAFeaturesCalculator

fc = MRNAFeaturesCalculator(df)
fc_df = fc.compute_mrna_features()
fc_df
```

```python
mrna_features_df = fc_df
mrna_features_df["protein_id"] = df["protein_id"]
mrna_features_df
```

```python
from modules.FeaturesCalculator import ProteinFeaturesCalculator

fc = ProteinFeaturesCalculator(df)
fc_df = fc.compute_protein_features()
fc_df
```

```python
protein_features_df = fc_df
protein_features_df["protein_id"] = df["protein_id"]
protein_features_df
```

```python
combined_features_df = pd.merge(mrna_features_df, protein_features_df, on="protein_id", how="left")
combined_features_df
```

```python
cols = ['protein_id'] + [col for col in combined_features_df.columns if col != 'protein_id']
combined_features_df = combined_features_df[cols]
combined_features_df
```

```python
combined_features_df.to_csv("biopython_features.csv", index=False)
```
